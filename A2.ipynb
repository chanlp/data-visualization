## Data Set
### New York Times Covid-19 dataset - Prison subset: https://github.com/nytimes/covid-19-data/tree/master/prisons

### Facility Dataset Focuses
__facility_type:__ The type of facility. <br>
__facility_city:__ The city where this facility is located. <br>
__facility_state:__ The state where the facility is located. <br>
__facility_lng and facility_lat:__ The longitude and latitude of the facility. <br>
__max_inmate_population_2020:__ The maximum number of inmates at the facility reported at any time from March 2020 through March 2021. <br>
__total_inmate_cases:__ Total number of cases of Covid-19 reported among inmates from the beginning of the pandemic through the end of March 2021. <br>
__total_inmate_deaths:__ Total number of inmates who were reported to have died of Covid-19 from the beginning of the pandemic through the end of March 2021. <br>
__total_officer_cases:__ Total number of cases of Covid-19 reported among correctional officers working at the facility from the beginning of the pandemic through the end of March 2021. <br>
__total_officer_deaths:__ Total number of correctional officers who worked at the facility who were reported to have died of Covid-19 from the beginning of the pandemic through the end of March 2021. 

### System Dataset Focuses
__system:__ The state of the prison system, or other system of detention facilities. For states, the data is reported only for state prisons, not for federal facilities or county jails within the state. <br>
__inmate_tests:__ The total number of P.C.R. tests conducted on inmates from the beginning of the pandemic through the end of March 2021. <br>
__total_inmate_cases:__ Total number of cases of Covid-19 reported among inmates in that system from the beginning of the pandemic through the end of March 2021. <br>
__total_inmate_deaths:__ Total number of inmates in that system who were reported to have died of Covid-19 from the beginning of the pandemic through the end of March 2021. <br>
__max_inmate_population_2020:__ The maximum number of inmates in the system from May 2020 through March 2021.
__total_officer_cases:__ Total number of cases of Covid-19 reported among correctional officers working in the system from the beginning of the pandemic through the end of March 2021. <br>
__total_officer_deaths:__ Total number of correctional officers working in the system who were reported to have died of Covid-19 from the beginning of the pandemic through the end of March 2021.

## Data Quality Assessment
- A primary issue with the quality of this dataset is that mamimum inmate population was not reported for every facility-type, and appears to be less likely to be reported for smaller facilities. 
- For this reason and others, only state prisons, federal prisons, and jails were included in this analysis. 
- Additionally, it is unlcear whether or not all facilities of a certain type are accounted for in this dataset, or if the total numbers of cases and tests per state remain consistent across the systems and facilities datasets. 
- For this reason, I largely kept analysis at the system and facility levels separate. When looking at geographic distributions, it was difficult to view all U.S. states and territories, so the entire analysis was restricted to the 48 contiguous U.S. states. 
- Additionally, some data fields contained NULL values. In this case, any data related to those fields was omitted. Overall, data quality is high and the precautions explained are likely sufficient for consistent and reliable analysis.

## Data Transformations Summary
 - __Choropleth maps:__ Divided total tests, cases, & deaths by the maximum state incarcerated population to get a proportion for each state's color encoding
 - __Pie charts:__ Totaled number of cases, death, and maximum population across facility type
 - __Bar Chart:__ Calculated the percent of inmate Covid-19 cases realtive to maximum facility population. i.e.  (cases/population)*100
 - __Packed Bubbles:__ Agreggated total number of cases for inmates and officers, per eacah facility type
 
All transformations were done directly in Tableau using the GUI
